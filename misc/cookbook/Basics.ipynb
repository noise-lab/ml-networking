## Basic Machine Learning with Scikit-Learn

Below is an example linear regression model, using Python's [scikit-learn](https://scikit-learn.org/) library.

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()

# Note: This example is very sloppy and just for demonstration right now.
# We did not split training and testing!!

# Train the Model
model = regr.fit(x.values.reshape(-1,1),y.values)

# Fit the Model
y_hat = regr.predict(np.array([0,3000]).reshape(-1,1))
```

```python
x = df1[' Total Backward Packets']
y = df1[' Total Fwd Packets']

plt.plot(x,y, '*')
plt.plot([0, 3000], y_hat, color='red', alpha=0.4, linewidth=3)

plt.xlabel('Total Packets')
plt.ylabel('Flow Duration')
plt.show()
```
